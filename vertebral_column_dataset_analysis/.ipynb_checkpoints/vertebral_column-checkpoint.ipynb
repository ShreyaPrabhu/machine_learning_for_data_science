### iii. Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
label_one_data = vertebral_data[vertebral_data["label"] == 1]
label_one_training = label_one_data.iloc[:140,:]
label_one_testing = label_one_data.iloc[140:,:]

label_zero_data = vertebral_data[vertebral_data["label"] == 0]
label_zero_training = label_zero_data.iloc[:70,:]
label_zero_testing = label_zero_data.iloc[70:,:]

training_data = label_one_training
training_data = training_data.append(label_zero_training)
training_features = training_data[["pelvic_incidence", "pelvic_tilt", "lumbar_lordosis_angle", "sacral_slope", "pelvic_radius", "degree_spondylolisthesis"]]
training_label = training_data[["label"]]

testing_data = label_one_testing
testing_data = testing_data.append(label_zero_testing)
testing_features = testing_data[["pelvic_incidence", "pelvic_tilt", "lumbar_lordosis_angle", "sacral_slope", "pelvic_radius", "degree_spondylolisthesis"]]
testing_label = testing_data[["label"]]
```

Following code is to track the least training error we have got and details regarding when we got the least values

```python
min_training_error = 1000000
min_training_error_reason = ""
```

### ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k for k  = {208; 205; ... ; 7; 4; 1}  (in reverse order). You are welcome to use smaller increments of k. Which k* is the most suitable k among those values? Calculate the confusion matrix, true positive rate, true negative rate, precision, and F1-score when k = k*

```python
train_error = {}
test_error = {}
k = 208
k_star = -1
k_star_accuracy = 0

while k > 0:
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = 2)
    knn_classifier.fit(training_features.to_numpy(), training_label.to_numpy())

    training_error = 1 - knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.to_numpy())
    testing_error = 1 - testing_accuracy
    
    train_error[k] = training_error
    test_error[k] = testing_error
    
    if testing_accuracy > k_star_accuracy or k_star == -1:
        k_star_accuracy = testing_accuracy
        k_star = k
    
    k = k - 3
    
keyval = min(train_error, key=train_error.get)
trainerror = train_error[keyval]
if trainerror < min_training_error or min_training_error == -1:
    min_training_error = train_error[keyval]
    min_training_error_reason = "Found this train error in C (ii) question with euclidean distance and k value " + str(keyval)
```

```python
plt.plot(*zip(*sorted(train_error.items())), label='training error')
plt.plot(*zip(*sorted(test_error.items())), label='testing error')
plt.gca().invert_xaxis()
plt.title("Training and Testing Error")
plt.xlabel("k value")
plt.ylabel("error value")
plt.legend(loc='upper right')
plt.show()
```

```python
print("k* value is : " + str(k_star))
print("Best error rate achieved with k* value is : " + str(1-k_star_accuracy))
print("\n")
knn_classifier = KNeighborsClassifier(n_neighbors=k_star, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = 2)
knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
testing_predictions = knn_classifier.predict(testing_features)

print("Confusion Matrix : ")
confusionMatrix = confusion_matrix(testing_label, testing_predictions)
print(confusionMatrix)

tn = confusionMatrix[0][0]
fn = confusionMatrix[1][0]
tp = confusionMatrix[1][1]
fp = confusionMatrix[0][1]


tpr = tp/(tp + fn)
print("True Positive Rate: " + str(tpr))
tnr = tn/(tn + fp)
print("True Negative Rate: " + str(tnr))
precision = tp/(tp + fp)
print("Precision: " + str(precision))
f1_score = (2*precision*tpr)/(precision + tpr)
print("F1-Score: " + str(f1_score))
```

### iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate, 3 which is obtained by some value of k, against the size of training set, when the size of training set is N = {10; 20; 30; : : : ; 210}. Note: for each N, select your training set by choosing the first N/3 rows of Class 0 and the first N - N/3 rows of Class 1 in the training set you created in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from {1; 6; 11; ... ; 196}. This plot is called a Learning Curve.

```python
best_k_values = {}
best_error_rate = {}

N = 10
while N <= 210:
    label_zero_subset = label_zero_training.iloc[:math.floor(N/3),:]
    label_one_subset = label_one_training.iloc[:N - math.floor(N/3),:]
    training_subset = label_zero_subset
    training_subset = training_subset.append(label_one_subset)
    training_subset_features = training_subset[["pelvic_incidence", "pelvic_tilt", "lumbar_lordosis_angle", "sacral_slope", "pelvic_radius", "degree_spondylolisthesis"]]
    training_subset_label = training_subset[["label"]]
    k = 1
    k_star = -1
    k_star_accuracy = 0

    while k <= N:
        knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = 2)
        knn_classifier.fit(training_subset_features.to_numpy(), training_subset_label.values.ravel())
        
        training_accuracy = knn_classifier.score(training_subset_features.to_numpy(), training_subset_label.to_numpy())
        training_error = 1 - training_accuracy
        
        if training_error < min_training_error or min_training_error == -1:
            min_training_error = training_error
            min_training_error_reason = "Found this train error in C (iii) question with euclidean distance and k value " + str(k) + " and training data size " + str(N)
        
        testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
        if testing_accuracy > k_star_accuracy or k_star == -1:
            k_star_accuracy = testing_accuracy
            k_star = k
    
        k = k + 5
    best_k_values[N] = k_star
    best_error_rate[N] = 1 - k_star_accuracy
    N = N + 10
```

```python
plt.plot(*zip(*sorted(best_error_rate.items())), label='best test error rate')
plt.title("Learning Curve Error Rate with varying N value")
plt.xlabel("N value")
plt.ylabel("best error rate")
plt.legend(loc='upper right')
plt.show()
```

## (d) Replace the Euclidean metric with the following metrics and test them. Summarize the test errors (i.e., when k = k*) in a table. Use all of your training data and select the best k when k = {1,6,.....196}

### i. Minkowski Distance:

#### A. which becomes Manhattan Distance with p = 1.

```python
k = 1
manhattan_k_star = -1
manhattan_k_star_accuracy = 0

while k <= 196:
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = 1)
    knn_classifier.fit(training_features.to_numpy(), training_label.to_numpy())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in D (i A) question with Manhattan distance and k value " + str(k)
        
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.to_numpy())
    if testing_accuracy > manhattan_k_star_accuracy or ((testing_accuracy == manhattan_k_star_accuracy) and manhattan_k_star == 1):
        manhattan_k_star_accuracy = testing_accuracy
        manhattan_k_star = k
    
    k = k + 5
print("k_star : " + str(manhattan_k_star))
print("Best error rate : " + str(1 - manhattan_k_star_accuracy))
    
```

#### B. with log10(p) = {0.1; 0.2; 0.3;.......1}. In this case, use the k  you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
exp = 0.1

log_p_exp = -1
log_p_exp_accuracy = 0
p = pow(10, exp)
while p <= 10:
    knn_classifier = KNeighborsClassifier(n_neighbors=manhattan_k_star, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = p)
    knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in D (i B) question with log10p distance and k value " + str(k)
        
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
    if testing_accuracy > log_p_exp_accuracy:
        log_p_exp_accuracy = testing_accuracy
        log_p_exp = exp
    
    exp = exp + 0.1
    p = pow(10, exp)

print("log_p_exp : " + str(log_p_exp))
print("Best error rate : " + str(1 - log_p_exp_accuracy))
```

#### C. which becomes Chebyshev Distance with p = infinity

```python
k = 1
chebyshev_k_star = -1
chebyshev_k_star_accuracy = 0

while k <= 196:
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'uniform', algorithm = 'brute', metric = 'minkowski', p = float('inf'))
    knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in D (i C) question with Chebyshev distance and k value " + str(k)
        
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
    if testing_accuracy > chebyshev_k_star_accuracy:
        chebyshev_k_star_accuracy = testing_accuracy
        chebyshev_k_star = k
    
    k = k + 5
print("k_star : " + str(chebyshev_k_star))
print("Best error rate : " + str(1 - chebyshev_k_star_accuracy))
```

### ii. Mahalanobis Distance

Check if the covariance of features matrix is invertible and set VI parameter for sklearn KNeighborsClassifier function

```python
def is_invertible(a):
    return a.shape[0] == a.shape[1] and np.linalg.matrix_rank(a) == a.shape[0]
```

```python
k = 1
mahalanobis_k_star = -1
mahalanobis_k_star_accuracy = 0

cov_matrix = training_features.cov()
vi_paramater = np.linalg.inv(training_features.cov())
if not is_invertible(cov_matrix):
    vi_paramater = np.linalg.pinv(training_features.cov())

while k <= 196:
    
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'uniform', algorithm = 'brute', metric = 'mahalanobis', metric_params={'VI': vi_paramater})
    knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in D (ii) question with mahalanobis distance and k value " + str(k)
        
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
    if testing_accuracy > mahalanobis_k_star_accuracy or ((testing_accuracy == mahalanobis_k_star_accuracy) and mahalanobis_k_star == 1):
        mahalanobis_k_star_accuracy = testing_accuracy
        mahalanobis_k_star = k
    
    k = k + 5
print("k_star : " + str(mahalanobis_k_star))
print("Best error rate : " + str(1 - mahalanobis_k_star_accuracy))
```

```python
knn_variants_summarization = pd.DataFrame(columns = ["distance_type", "k_star", "lop_p_exp_value", "error", "error_percent"])
knn_variants_summarization = knn_variants_summarization.append({'distance_type': "manhattan", 'k_star': manhattan_k_star, 'error': 1 - manhattan_k_star_accuracy, 'error_percent': (1 - manhattan_k_star_accuracy)*100}, ignore_index=True)
knn_variants_summarization = knn_variants_summarization.append({'distance_type': "log_p_exp", 'lop_p_exp_value': log_p_exp, 'error': 1 - log_p_exp_accuracy, 'error_percent': (1 - log_p_exp_accuracy)*100}, ignore_index=True)
knn_variants_summarization = knn_variants_summarization.append({'distance_type': "chebyshev", 'k_star': chebyshev_k_star, 'error': 1 - chebyshev_k_star_accuracy, 'error_percent': (1 - chebyshev_k_star_accuracy)*100}, ignore_index=True)
knn_variants_summarization = knn_variants_summarization.append({'distance_type': "mahalanobis", 'k_star': mahalanobis_k_star, 'error': 1 - mahalanobis_k_star_accuracy, 'error_percent': (1 - mahalanobis_k_star_accuracy)*100}, ignore_index=True)
knn_variants_summarization.head(4)
```

## (e) The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater in influence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k = {1,6, 11, .....196}

```python
k = 1
euclidean_k_star = -1
euclidean_k_star_accuracy = 0

while k <= 196:
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'distance', algorithm = 'brute', metric = 'minkowski', p = 2)
    knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in E question with euclidean weighted distance and k value " + str(k)
        
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
    if testing_accuracy > euclidean_k_star_accuracy or ((testing_accuracy == euclidean_k_star_accuracy) and euclidean_k_star == 1):
        euclidean_k_star_accuracy = testing_accuracy
        euclidean_k_star = k
    
    k = k + 5
    
print("k_star : " + str(euclidean_k_star))
print("Best error rate : " + str(1 - euclidean_k_star_accuracy))
```

```python
k = 1
chebyshev_k_star = -1
chebyshev_k_star_accuracy = 0

while k <= 196:
    knn_classifier = KNeighborsClassifier(n_neighbors=k, weights = 'distance', algorithm = 'brute', metric = 'minkowski', p = float('inf'))
    knn_classifier.fit(training_features.to_numpy(), training_label.values.ravel())
    
    training_accuracy = knn_classifier.score(training_features.to_numpy(), training_label.to_numpy())
    training_error = 1 - training_accuracy
    if training_error < min_training_error or min_training_error == -1:
        min_training_error = training_error
        min_training_error_reason = "Found this train error in E question with chebyshev weighted distance and k value " + str(k)
        
    
    
    testing_accuracy = knn_classifier.score(testing_features.to_numpy(), testing_label.values.ravel())
    if testing_accuracy > chebyshev_k_star_accuracy or ((testing_accuracy == chebyshev_k_star_accuracy) and chebyshev_k_star == 1):
        chebyshev_k_star_accuracy = testing_accuracy
        chebyshev_k_star = k
    
    k = k + 5
    
print("k_star : " + str(chebyshev_k_star))
print("Best error rate : " + str(1 - chebyshev_k_star_accuracy))
```
