```python
bending_testing = ["dataset1.csv", "dataset2.csv"]
others_testing = ["dataset1.csv", "dataset2.csv", "dataset3.csv"]

colnames=['time', 'avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23', 'action_instance']
training_data = pd.DataFrame(columns = colnames)
testing_data = pd.DataFrame(columns = colnames)

file_list = os.listdir(bending_one)
for file in file_list:
    data = pd.read_csv(bending_one + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "bending1" + "_" + file
    if file in bending_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(bending_two)
for file in file_list:
    if file == "dataset4.csv":
        data = pd.read_csv(bending_two + file, names=colnames, delimiter = ' ', skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    else:
        data = pd.read_csv(bending_two + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "bending2" + "_" + file
    if file in bending_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(cycling)
for file in file_list:
    data = pd.read_csv(cycling + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "cycling" + "_" + file
    if file in others_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(lying)
for file in file_list:
    data = pd.read_csv(lying + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "lying" + "_" + file
    if file in others_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(sitting)
for file in file_list:
    data = pd.read_csv(sitting + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "sitting" + "_" + file
    if file in others_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(standing)
for file in file_list:
    data = pd.read_csv(standing + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "standing" + "_" + file
    if file in others_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
        
file_list = os.listdir(walking)
for file in file_list:
    data = pd.read_csv(walking + file, names=colnames, skiprows = 5, skip_blank_lines = True, error_bad_lines = False, warn_bad_lines = True, header = None) 
    data["action_instance"] = "walking" + "_" + file
    if file in others_testing:
        testing_data = testing_data.append(data)
    else:
        training_data = training_data.append(data)
```

### (ii) Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them directly

```python
'avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23'

colnames=['min_1', 'max_1', 'mean_1', 'median_1', 'stddev_1', 'first_quartile_1', 'third_quartile_1', 
          'min_2', 'max_2', 'mean_2', 'median_2', 'stddev_2', 'first_quartile_2', 'third_quartile_2',
          'min_3', 'max_3', 'mean_3', 'median_3', 'stddev_3', 'first_quartile_3', 'third_quartile_3',
          'min_4', 'max_4', 'mean_4', 'median_4', 'stddev_4', 'first_quartile_4', 'third_quartile_4',
          'min_5', 'max_5', 'mean_5', 'median_5', 'stddev_5', 'first_quartile_5', 'third_quartile_5',
          'min_6', 'max_6', 'mean_6', 'median_6', 'stddev_6', 'first_quartile_6', 'third_quartile_6',
          'instance']
time_domain_features = pd.DataFrame(columns = colnames)

all_data = training_data
all_data = all_data.append(testing_data)

instances = sorted(all_data['action_instance'].unique())
for instance in instances:
    df_instance = all_data[all_data['action_instance'] == instance]
    time_domain_features = time_domain_features.append({'instance': instance, 
                                 'min_1' : df_instance['avg_rss12'].min(),
                                 'max_1' : df_instance['avg_rss12'].max(),
                                 'mean_1' : df_instance['avg_rss12'].mean(),
                                 'median_1' : df_instance['avg_rss12'].median(),
                                 'stddev_1' : df_instance['avg_rss12'].std(),
                                 'first_quartile_1' : df_instance['avg_rss12'].quantile(0.25),
                                 'third_quartile_1' : df_instance['avg_rss12'].quantile(0.75),
                                 'min_2' : df_instance['var_rss12'].min(),
                                 'max_2' : df_instance['var_rss12'].max(),
                                 'mean_2' : df_instance['var_rss12'].mean(),
                                 'median_2' : df_instance['var_rss12'].median(),
                                 'stddev_2' : df_instance['var_rss12'].std(),
                                 'first_quartile_2' : df_instance['var_rss12'].quantile(0.25),
                                 'third_quartile_2' : df_instance['var_rss12'].quantile(0.75),
                                 'min_3' : df_instance['avg_rss13'].min(),
                                 'max_3' : df_instance['avg_rss13'].max(),
                                 'mean_3' : df_instance['avg_rss13'].mean(),
                                 'median_3' : df_instance['avg_rss13'].median(),
                                 'stddev_3' : df_instance['avg_rss13'].std(),
                                 'first_quartile_3' : df_instance['avg_rss13'].quantile(0.25),
                                 'third_quartile_3' : df_instance['avg_rss13'].quantile(0.75),
                                 'min_4' : df_instance['var_rss13'].min(),
                                 'max_4' : df_instance['var_rss13'].max(),
                                 'mean_4' : df_instance['var_rss13'].mean(),
                                 'median_4' : df_instance['var_rss13'].median(),
                                 'stddev_4' : df_instance['var_rss13'].std(),
                                 'first_quartile_4' : df_instance['var_rss13'].quantile(0.25),
                                 'third_quartile_4' : df_instance['var_rss13'].quantile(0.75),
                                 'min_5' : df_instance['avg_rss23'].min(),
                                 'max_5' : df_instance['avg_rss23'].max(),
                                 'mean_5' : df_instance['avg_rss23'].mean(),
                                 'median_5' : df_instance['avg_rss23'].median(),
                                 'stddev_5' : df_instance['avg_rss23'].std(),
                                 'first_quartile_5' : df_instance['avg_rss23'].quantile(0.25),
                                 'third_quartile_5' : df_instance['avg_rss23'].quantile(0.75),
                                 'min_6' : df_instance['var_rss23'].min(),
                                 'max_6' : df_instance['var_rss23'].max(),
                                 'mean_6' : df_instance['var_rss23'].mean(),
                                 'median_6' : df_instance['var_rss23'].median(),
                                 'stddev_6' : df_instance['var_rss23'].std(),
                                 'first_quartile_6' : df_instance['var_rss23'].quantile(0.25),
                                 'third_quartile_6' : df_instance['var_rss23'].quantile(0.75),
                                }, ignore_index=True)
```

### (iii) Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Pythons bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.

```python
time_domain_features.std()
```

```python
colnames=['min_1', 'max_1', 'mean_1', 'median_1', 'stddev_1', 'first_quartile_1', 'third_quartile_1', 
          'min_2', 'max_2', 'mean_2', 'median_2', 'stddev_2', 'first_quartile_2', 'third_quartile_2',
          'min_3', 'max_3', 'mean_3', 'median_3', 'stddev_3', 'first_quartile_3', 'third_quartile_3',
          'min_4', 'max_4', 'mean_4', 'median_4', 'stddev_4', 'first_quartile_4', 'third_quartile_4',
          'min_5', 'max_5', 'mean_5', 'median_5', 'stddev_5', 'first_quartile_5', 'third_quartile_5',
          'min_6', 'max_6', 'mean_6', 'median_6', 'stddev_6', 'first_quartile_6', 'third_quartile_6',
          'instance']

print("90% Bootstrap confidence interval")
rng = np.random.default_rng()
for column_name in colnames:
    if column_name == "instance":
        continue
    data = time_domain_features[[column_name]]
    data = (data,)
    result = bootstrap(data, np.std, confidence_level=0.9, random_state=rng, method = 'basic')
    print(column_name + " : " + str(result.confidence_interval))
```

## I collect a set of data (n = 100 observations) containing a single predictor and a quantitative response. I then fit a linear regression model to the data, as well as a separate cubic regression, i.e. Y = 0 + 1X + 2X2 + 3X3 + .

## (a) Suppose that the true relationship between X and Y is linear, i.e. Y = 0 + 1X + . Consider the training residual sum of squares (RSS) for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

Training RSS with cubic regression will be lower

We will expect to have lower training RSS with cubic regression model (Y = 0 + 1X + 2X2 + 3X3 + ). This is because the cubic regression model would be more flexible and hence would have overfit the training dataset. 

## (b) Answer (a) using test rather than training RSS.

Test RSS for cubic regression will be higher

The cubic regression model has overfit the data set during training. This leads to the model to possibly not perform well on test dataset. Hence, the cubic regression will have higher RSS on test dataset when compared.

## (c) Suppose that the true relationship between X and Y is not linear, but we dont know how far it is from linear. Consider the training RSS for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

Training RSS with cubic regression will be lower

If the ground true relationship between X and Y is not known, then, if we fit linear and cubic regression models, the training RSS will be lower for cubic regression as it is more flexible and will be tightly fit to the training dataset. On the other hand, a linear model most probably will be underfit (if the dataset has polynomial relation) and hence is likely to have higher RSS

## (d) Answer (c) using test rather than training RSS.

There is not enough data about true relationship to comment on and compare the test RSS of linear and cubic regression models. 

Based on true relationship, it depends on which model was better fit and how well it would be able to generalize on test dataset.
